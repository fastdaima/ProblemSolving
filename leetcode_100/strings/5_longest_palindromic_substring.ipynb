```python
class Solution:
    def expand_in_middle(self,s:str,left:int, right:int)->str:
        while (left>=0 and right <len(s) and s[left]==s[right]):
            left -=1
            right +=1
        return s[left+1:right]
```

```python
    def longestPalindrome(self,s:str)->str:
        out = ""
        odd_outputs = ""
        even_outputs = ""
        for ptr in range(0,len(s)):
            odd  = self.expand_in_middle(s, ptr-1, ptr+1)
            even = self.expand_in_middle(s, ptr, ptr+1)
            if(len(odd)>len(odd_outputs)):  odd_outputs = odd
            if(len(even)>len(even_outputs)): even_outputs = even
        out = max(odd_outputs, even_outputs,key=len)
        return out
```

```python
s = Solution()
print (s.longestPalindrome("abbba"))
print (s.longestPalindrome("abbba"))
print (s.longestPalindrome("aba"))
print (s.longestPalindrome("abbbba"))
```
